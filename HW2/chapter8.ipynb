### Aproximation Contours

A taylor series is used to approximate the function. we made second order approximations
\begin{equation}
F(X) = F(X^*) + \nabla F(X)^T |_{X=X*} (X-X*) + \frac{1}{2} (X-X^*)^T \nabla^2 F(X) |_{X-X^*} (X-X^*)
\end{equation}
And X for the question is
$$
X = \begin{pmatrix}
x_1 \\
x_2
\end{pmatrix}
$$

```python
def taylor_second_order(x1, x2, point, A, val):
    """
    second order approximation function

    INPUTS:
    -------
    x1, x2: numpy arrays
    point:  is the point we are going to approximate
    A:  hessian matrix
    val:  the value for the function, F(X^*)

    OUTPUTS:
    ----------
    the approximated value
    """
        
    p1 = x1 - point[0]
    p2 = x2 - point[1]

    part = p1 * (A[0,0] + p1 + A[0, 1] * p2) + p2 * (A[1, 0] * p2 + A[1,1] * p1) 

    return val + part
```

$$
X = \begin{pmatrix}
-0.6504 \\
-0.6504
\end{pmatrix}
$$

```python
x1 = np.linspace(-5, 5, 100)
x2 = np.linspace(-5, 5, 100)
x2 = x2.reshape(-1, 1)

point = [-0.6504, -0.6504]
A = np.matrix('20.43 8.43; 8.43 20.43') 
val = 6.076
y = taylor_second_order(x1, x2, point, A, val)
plt.contour(y)
plt.show()
```
